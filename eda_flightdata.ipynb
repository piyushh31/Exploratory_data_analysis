```python
encoding_list = []
for i in df:
     if len(df[i].unique())<=2 and df[i].dtype =='object':
        encoding_list.append(i)
dict1 = {}
for i in encoding_list:
    dict1[i] = list(pd.unique(df[i]))
print(dict1)
for i in dict1.keys():
    for j in dict1[i]:
        df[i].replace(j,dict1[i].index(j),inplace = True)
```

```python
# One hot encoding (Dummy Variables)
dummy_var = []
for i in df:
    if df[i].dtype == 'object':
        dummy_var.append(i)
print(dummy_var)
for i in dummy_var:
    temp_df = pd.get_dummies(df[i])
    del temp_df[temp_df.columns[-1]] #Removing Multicollinearlity
    df.drop(i,axis = 1,inplace = True)
    df = pd.concat([df,temp_df],axis='columns')
```

```python
loadings_df = pd.DataFrame(fa.loadings_,index = df.columns)
loadings_df
```

```python
#Finding out the results as per maximum loadings
factors_dict = {}
for i in loadings_df:
    factors_dict[f'Factor {i}'] = []
print(factors_dict)
for j in list(loadings_df.index):
    temp = []
    for k in list(loadings_df.loc[j,:]): 
        temp.append(abs(k))
    factors_dict[f'Factor {temp.index(max(temp))}'].append(j)
```
